## Objective:
Write a program that prints the numbers from 1 to 100. But, for multiples of three print **Fizz** instead of the number, similarly, for multiples of five print **Buzz**. Finally, for numbers which are multiples of both three and five, print **FizzBuzz**

## Programming language:
<ul>
    <li>Python</li>
</ul>

## Libraries

For this project there were no use of external libraries.

## Code:

```python
# We stablish a for loop that will iterate from 1 to 100.
for x in range(1,101):
    # Note that this must be the first condition.
    if (x % 3 == 0) & (x % 5 == 0):
        print("FizzBuzz")
    elif x % 3 == 0:
        print("Fizz")
    elif x % 5 == 0:
        print("Buzz")
    else:
        print(x)
```
